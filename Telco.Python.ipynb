```python
# let's run an Independent T-test on total charges and Churn value. 
ttest = customers.filter(["Monthly Charges", "Churn Value"], axis=1)
```

```python
ttest.head()
```

```python
ttest.rename(columns = {'Monthly Charges':'Monthly_Charges', 'Churn Value':'Churn_Value'}, inplace = True)
```

```python
ttest.info()
```

```python
# First we test assumptions:
ttest.Churn_Value[ttest.Monthly_Charges].hist()
```

```python
# Then we run the analysis
ttest_ind(ttest.Monthly_Charges[ttest.Churn_Value == '0'], ttest.Monthly_Charges[ttest.Churn_Value == '1'])
```

```python
# It would seem our efforts to produce an independent t-test with either the "Monthly Charges" or the "Total Charges" columns remains fruitless. 
# We shall now attempt a single-sample t-test on the data.

# start by getting the mean value in monthly charges column
ttest.Monthly_Charges.mean()
```

```python
# now let's add a value to see how probable it is
stats.ttest_1samp(ttest['Monthly_Charges'], 65)
```

```python
# now a second lower value
stats.ttest_1samp(ttest['Monthly_Charges'], 63)
```

```python
# It becomes clear that our mean is very indicative of what most people paid. Any deviation from that mean, even by just a couple dollars at the most, reveals a statistically significant probability.
```
